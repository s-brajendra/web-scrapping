```python
import requests
from bs4 import BeautifulSoup
import json
import numpy as np
import pandas as pd

response = requests.get("https://books.toscrape.com/catalogue/page-1.html")
data = response.text


```

```python
data = BeautifulSoup(data)
```

```python
i = 1
urls = []
values = []
j = 1
while(data.find(class_='next') != None):
    link = 'https://books.toscrape.com/catalogue/' + data.find(class_ = "next").a["href"]
    urls.append(link)
    print("currently on page: ",i)
    
    
    
    
    
    books_card = data.find_all(class_ = "product_pod")
    for card in books_card:
        book = []
        image_link = "https://books.toscrape.com/catalogue/" +  card.div.a["href"]
        book_name = card.h3.a["title"]
        book_price = card.find(class_= "price_color").string[1:]
        availability = books_card[0].find(class_ = "instock availability").get_text().split("\n")[3].strip()
        book.append(book_name.strip())
        book.append(book_price)
        book.append(availability)
        book.append(image_link)
        
        values.append(book)
        print("got book",j)
        j = j+1
        
        
        
        
    
    
    
    
    
    
    
    
    
    # for going to next page
    response = requests.get(link)
    data = response.text
    data = BeautifulSoup(data)
    i = i+1
        
```

```python
df = pd.DataFrame(values,columns = ["Name","price","availability","BookImg"]) 
df
```
